# SMP

```python
from threading import Thread, Lock
from time import sleep

def task(lock):
    # Espera inicial para dar tiempo al main thread de adquirir el lock
    sleep(0.5)
    # Bucle activo (spinlock)
    while True:
        acquired = lock.acquire(blocking=False)  # Intento no bloqueante
        if acquired:
            print("Hilo secundario adquirió el lock")
            lock.release()
            break
        else:
            print("Esperando...")  # Opcional: mensaje de depuración

# Crear lock y ejecutar
lock = Lock()

# Main thread adquiere el lock primero
with lock:
    print("Main thread adquirió el lock")
    hilo = Thread(target=task, args=(lock,))
    hilo.start()
    sleep(3)  # Simula trabajo en sección crítica
    print("Main thread libera el lock")

hilo.join()
```
